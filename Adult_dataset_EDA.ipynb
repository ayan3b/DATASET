```python
# a funtion to assign "undefined" where value '?'
def assign_unknow(val):
  if val == '?':
    return('undefined')
  else :
    return val
print(assign_unknow('abc'))
```

```python
df['workclass'] = df.workclass.apply(lambda x :assign_unknow(x))

df.workclass.value_counts()
```

```python
cat_feature = ['education']
#'workclass','education','marital-status','occupation','relationship','race','sex','native-country']
for f in cat_feature:
  group_feature = [f,'income']
  out_df = df[group_feature].groupby(group_feature).agg({'income': 'count'})
  print('Starting of the feature :', f)
  print(out_df)
```

```python
# a funtion to reduce the education class

def assign_education_class(val):
  if val == '9th' or val == '7th-8th' or val == '5th-6th' or val == '1st-4th' or val == '10th':
    return('School')
  elif val == '11th' or val == '12th' or val == 'HS-grad':
    return('HS-grad')
  elif val == 'Some-college' or val == 'Bachelors' :
    return('Bachelors')
  elif val == 'Assoc-acdm' or val == 'Assoc-voc' :
    return('AssociateLevel')
  else :
    return val
print(assign_education_class('9th'))
```

```python
df['education'] = df.education.apply(lambda x : assign_education_class(x))

df.education.value_counts()
```
